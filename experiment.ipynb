```python
import requests
import json

def get_output(input_text):
    api_url = "http://localhost:11434/api/generate"
    payload = {
        "model": "mistral:7b",
        "prompt": input_text,
        "format": "json",
        "stream": False
    }
    try:
        response = requests.post(api_url, data=json.dumps(payload))
        if response.status_code == 200:
            print("Request was successful!")
            return response.json()
        else:
            print(f"Failed with status code {response.status_code}")
            return response.text
    except Exception as e:
        print(f"An error occurred: {e}")
    return None

get_output("Given that the word is used in steel industries, translate the following word from german to english: ungebeizt, nicht geglüht")
```

```python
file_name = "./resources\source2.xlsx"
sheet_name = "First choice "
df = pd.read_excel(file_name, sheet_name=sheet_name)
df
```

```python
df1 = df.iloc[1:12]
df1.columns = df.iloc[0].tolist()
df1.reset_index(drop=True, inplace=True)
df1
```

```python
df2 = df.iloc[15:23]
df2.columns = df.iloc[14].tolist()
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df3 = df.iloc[26:]
df3.columns = df.iloc[25].tolist()
df3.reset_index(drop=True, inplace=True)
df3 = df3.dropna(axis=1, how='all')
df3
```

```python
# cleaning
df = pd.concat([df1, df2, df3])
df["Source"] = file_name
df["Sheet Name"] = sheet_name
df
```
