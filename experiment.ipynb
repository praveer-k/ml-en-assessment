```python
import pandas as pd
from glob import glob

for file in glob("./resources/*.xlsx"):
    xls = pd.ExcelFile(file)
    for sheet_name in xls.sheet_names:
        print(f"file: {file}, sheet: {sheet_name}")
```

```python
strength = ['RP02', 'RM', 'A', 'STA']
composition = ['AG', 'Al', 'Ars', 'B', 'C', 'Ca', 'Cr', 'S', 'Cu', 'Mn', 'Mo', 'N', 'Nb', 'Ni', 'P', 'Si', 'Sn', 'Ti', 'V', 'Zr']

def set_properties(row):
    def kv(row, keys):
        parts = []
        for key in keys:
            parts.append(f"{key}: {row[key]}")
        return "\n\t".join(parts)    
    props = [f"Grade: {row['Grade']}", f"Strength: \n\t{kv(row, strength)}", f"Chemical composition: \n\t{kv(row, composition)}"]
    return "\n".join(props)

def set_description(row):
    keys = ['Description', 'Finish']
    parts = []
    for key in keys:
        parts.append(f"{key}: {row[key]}")
    return "\n".join(parts)

file_path = "./resources/source1.xlsx"
sheet_name = "Sheet1"
df = pd.read_excel(file_path, sheet_name=sheet_name)

# cleaning
df["Finish"] = df["Finish"].str.replace("ungebeizt, nicht geglÃ¼ht", "Unpickled, Not Annealed")
df["Finish"] = df["Finish"].str.replace("gebeizt und geglÃ¼ht", "Pickled and Annealed")

df["Description"] = df["Description"].str.replace('Sollmasse (Gewicht) unterschritten', "Below target weight")
df["Description"] = df["Description"].str.replace('Kantenfehler - FS-Kantenrisse', "Edge defects - FS edge cracks")
df["Description"] = df["Description"].str.replace('LÃ¤ngs- oder Querrisse', "Longitudinal or transverse cracks")

df["properties"] = df.apply(set_properties, axis=1)
df["Description"] = df.apply(set_description, axis=1)
df.drop(columns=["Finish"] + strength + composition, inplace=True)
df["file_path"] = file_path
df["sheet_name"] = sheet_name

df.columns = [col.lower().strip().replace(" ","_") for col in df.columns]
df.rename(columns={"quality/choice": "choice", "grade": "material_name", "thickness_(mm)": "height", "width_(mm)": "breadth", "gross_weight_(kg)": "weight"}, inplace=True)
df["choice"] = df["choice"].str.replace("2nd", "2")
df["choice"] = df["choice"].fillna("-1")
df["choice"] = df["choice"].astype(int)

df['length'] = None
df["quantity_unit"] = "piece"
df["dimension_unit"] = "mm" 
df["weighing_unit"] = "kg"

df["quantity"] = None
df["material_id"] = None
df["total_price"] = None
df["price_per_unit"] = None
df["supplier"] = None
df["reserved"] = None

df = df.reindex(columns=['material_id', 'material_name', 'quantity', 'quantity_unit', 'price_per_unit', 'supplier', 'length', 'breadth', 'height', 'dimension_unit', 'weight', 'weighing_unit', 'properties', 'description', 'choice', 'reserved'])
df
```

```python
import re

pattern = r'(?P<length>\d{1,2}(?:[,.]\d{2})?) x (?P<breadth>\d{1,4}(?:[,.]\d{2})?)(?: x (?P<height>\d{1,4}(?:[,.]\d{2})?))?(?: (?P<dimension_unit>mm))?'
    
def extract_dimensions(text):
    match = re.search(pattern, text)
    if match:
        return match.groupdict()
    return {"length": None, "breadth": None, "height": None, "dimension_unit": None}
```

```python
import pandas as pd

file_path = "./resources/source2.xlsx"
sheet_name = "First choice "
df = pd.read_excel(file_path, sheet_name=sheet_name)
# df1
df1 = df.iloc[1:12]
df1.columns = df.iloc[0].tolist()
df1.reset_index(drop=True, inplace=True)
# df2
df2 = df.iloc[15:23]
df2.columns = df.iloc[14].tolist()
df2.reset_index(drop=True, inplace=True)
# df3
df3 = df.iloc[26:]
df3.columns = df.iloc[25].tolist()
df3.reset_index(drop=True, inplace=True)
df3 = df3.dropna(axis=1, how='all')
# cleaning
df = pd.concat([df1, df2, df3])
df.columns = [col.lower().strip().replace(" ","_") for col in df.columns]
df.rename(columns={"article_id": "material_id", "material": "properties", "description": "other_description", "opmerking": "description"}, inplace=True)
df["description"] = df["description"].fillna("")
df["properties"] = df.apply(lambda row: row["properties"] + " Oiled" if str(row["other_description"]).strip()=="Material is Oiled" else row["properties"], axis=1)
df.drop(columns=["other_description"], inplace=True)
df["properties"] = df["properties"].str.replace(r'(\d)x', r'\1 x ', regex=True)
df = df.join(df["properties"].apply(extract_dimensions).apply(pd.Series))
df["dimension_unit"] = df["dimension_unit"].fillna("mm")
df["properties"] = df["properties"].str.replace(r'(\w)\+', r'\1 +', regex=True)
df["properties"] = df["properties"].str.replace(r'(\d)mm', r'\1 mm', regex=True)
df["properties"] = df["properties"].str.replace(pattern, '', regex=True)
df["properties"] = df["properties"].apply(lambda x: x.replace("geolied", "Oiled"))
df["quantity_unit"] = "piece"
df["total_price"] = None
df["price_per_unit"] = None
df["supplier"] = None
df["weighing_unit"] = None
df["choice"] = 1
df = df.reindex(columns=['material_id', 'material_name', 'quantity', 'quantity_unit', 'price_per_unit', 'supplier', 'length', 'breadth', 'height', 'dimension_unit', 'weight', 'weighing_unit', 'properties', 'description', 'choice', 'reserved'])
df["file_path"] = file_path.replace('\\', '/')
df["sheet_name"] = sheet_name.strip()
df
```
